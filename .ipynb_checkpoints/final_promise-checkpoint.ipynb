# Time Spent on Each Step: Average duration users spend on each step.

```python
# Convert the 'Login_time' column to datetime type
df_web_all_2['date_time'] = pd.to_datetime(df_web_all_2['date_time'])

# Sort the dataframe by 'ID_client' and 'Login_time'
df_web_all_2 = df_web_all_2.sort_values(by=['visit_id', 'date_time'])

# Calculate the time spent on each step for each client
df_web_all_2['Time_spent'] = df_web_all_2.groupby('visit_id')['date_time'].diff()
```

```python
# Filter out rows with NaT values in the 'Time_spent' column
filtered_df = df_web_all_2.dropna(subset=['Time_spent'])

# Convert timedelta to seconds
filtered_df['Time_spent_seconds'] = filtered_df['Time_spent'].dt.total_seconds()
```

```python
Q1 = filtered_df['Time_spent'].quantile(0.25)
Q3 = filtered_df['Time_spent'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for the outliers
lower_bound = Q1 - 10 * IQR
upper_bound = Q3 + 10 * IQR

# Identify and remove the outliers
filtered_df = filtered_df[filtered_df['Time_spent'] < upper_bound]
```

```python
filtered_df
```

* KEEP ONLY THOSE VISIT_IDs THAT HAVE THE FOUR STEPS ONE AFTER ANOTHER

```python
# Step 1: Group the DataFrame by 'visit_id'
grouped = filtered_df.groupby('visit_id')

# Step 2: Check if each group contains all four steps
def has_required_steps(group):
    return set(group['process_step']) == {'step_1', 'step_2', 'step_3', 'confirm'}

# Step 3: Filter groups to keep only those that have all four required steps
filtered_groups = grouped.filter(has_required_steps)

# Step 4: Extract corresponding rows from the original DataFrame
filtered_clients = filtered_df[filtered_df['visit_id'].isin(filtered_groups['visit_id'])]

# Display the resulting DataFrame
filtered_clients
```

* REMOVE REPETED STEPS FOR SAME CLIENT_ID

```python
# Sort the DataFrame by 'date_time' within each group
filtered_df_sorted = filtered_df.sort_values(by='date_time')

# Step 1: Group the DataFrame by 'visit_id'
grouped = filtered_df_sorted.groupby('visit_id')

# Step 2: Define a function to filter each group
def filter_group(group):
    # Keep only the earliest occurrence of each step
    filtered_group = group.drop_duplicates(subset='process_step')
    # Check if the group contains all four required steps
    if set(filtered_group['process_step']) == {'step_1', 'step_2', 'step_3', 'confirm'}:
        return filtered_group
    else:
        return pd.DataFrame()  # Return an empty DataFrame if the group does not meet the criteria

# Step 3: Apply the filtering function to each group and concatenate the results
filtered_clients = grouped.apply(filter_group).reset_index(drop=True)

# Display the resulting DataFrame
filtered_clients
```

# Completion Rate: proportion of users who reach the final ‘confirm’ step.

```python
# Assuming your dataframe is named filtered_df
# Assuming 'process_step' values are limited to start, step_1, step_2, step_3, and confirm
# Create a new column 'step_number' based on the logic you provided
filtered_df['step_number'] = filtered_df['process_step'].map({'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4})
# Create a pivot table
pivot_df = filtered_df.pivot_table(index='visit_id', columns='step_number', values='process_step', aggfunc='first')
# Rename the columns to match the step numbers
pivot_df.columns = [f'step_{col}' for col in pivot_df.columns]
# Reset index to make 'visit_id' a column again
pivot_df = pivot_df.reset_index()
# Merge the pivoted DataFrame back to the original DataFrame based on 'visit_id'
merged_df = pd.merge(filtered_df, pivot_df, how='left', on='visit_id')
# Print the resulting DataFrame
merged_df
```

```python
# Assuming your dataframe is named merged_df
# Assuming the columns are step_0, step_1, step_2, step_3, step_4
# List of columns to check for NaN
columns_to_check = ['step_0','step_4']
# Create a new column with True if all specified columns are not NaN, False otherwise
merged_df['all_steps_completed'] = merged_df[columns_to_check].notna().all(axis=1)
# Print the resulting DataFrame
merged_df
```

# Error Rate

** True = error

```python
# Assuming df_web_all2 is your DataFrame
# Step 1: Sort the DataFrame
df_web_all_2.sort_values(by=['visit_id', 'date_time'], inplace=True)

# Step 2: Identify instances of going back to a previous step
df_web_all_2['previous_step'] = df_web_all_2.groupby(['visit_id'])['process_step'].shift(1)

# Define the correct order of steps
step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Function to check if the previous step is in the correct order or if it's the same as the current step
def check_step_order(previous_step, current_step):
    if pd.isna(previous_step):
        return current_step != 'start'  # Consider NaN as error except when current step is 'start'
    elif previous_step == current_step:
        return True  # Return True if previous step is the same as current step
    else:
        previous_index = step_order.index(previous_step)
        current_index = step_order.index(current_step)
        return previous_index > current_index  # True if previous step index is greater than current step index

# Apply the function to create the 'column_error' column
df_web_all_2['column_error'] = df_web_all_2.apply(lambda row: check_step_order(row['previous_step'], row['process_step']), axis=1)

# Display the dataframe
df_web_all_2
```

```python
test_errors = df_web_all_2[df_web_all_2['Variation'] == 'Test']['column_error'].sum()
test_total = len(df_web_all_2[df_web_all_2['Variation'] == 'Test'])
test_error_rate = test_errors / test_total

control_errors = df_web_all_2[df_web_all_2['Variation'] == 'Control']['column_error'].sum()
control_total = len(df_web_all_2[df_web_all_2['Variation'] == 'Control'])
control_error_rate = control_errors / control_total

# Create a bar plot
plt.bar(['Control','Test'], [control_error_rate,test_error_rate], color=['blue', 'red'])
plt.xlabel('Group')
plt.ylabel('Error Rate')
plt.title('Error Rates in Test vs Control')
plt.show()
```
