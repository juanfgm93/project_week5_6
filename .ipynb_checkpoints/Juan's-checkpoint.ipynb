# Time Spent on Each Step: Average duration users spend on each step.

```python
# Convert the 'Login_time' column to datetime type
df_web_all_2['date_time'] = pd.to_datetime(df_web_all_2['date_time'])

# Sort the dataframe by 'ID_client' and 'Login_time'
df_web_all_2 = df_web_all_2.sort_values(by=['visit_id', 'date_time'])

# Calculate the time spent on each step for each client
df_web_all_2['Time_spent'] = df_web_all_2.groupby('visit_id')['date_time'].diff()
```

```python
# Filter out rows with NaT values in the 'Time_spent' column
filtered_df = df_web_all_2.dropna(subset=['Time_spent'])

# Convert timedelta to seconds
filtered_df['Time_spent_seconds'] = filtered_df['Time_spent'].dt.total_seconds()
```

```python
Q1 = filtered_df['Time_spent'].quantile(0.25)
Q3 = filtered_df['Time_spent'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for the outliers
lower_bound = Q1 - 10 * IQR
upper_bound = Q3 + 10 * IQR

# Identify and remove the outliers
filtered_df = filtered_df[filtered_df['Time_spent'] < upper_bound]
```

```python
filtered_df
```

* KEEP ONLY THOSE VISIT_IDs THAT HAVE THE FOUR STEPS ONE AFTER ANOTHER

```python
# Step 1: Group the DataFrame by 'visit_id'
grouped = filtered_df.groupby('visit_id')

# Step 2: Check if each group contains all four steps
def has_required_steps(group):
    return set(group['process_step']) == {'step_1', 'step_2', 'step_3', 'confirm'}

# Step 3: Filter groups to keep only those that have all four required steps
filtered_groups = grouped.filter(has_required_steps)

# Step 4: Extract corresponding rows from the original DataFrame
filtered_clients = filtered_df[filtered_df['visit_id'].isin(filtered_groups['visit_id'])]

# Display the resulting DataFrame
filtered_clients
```

* REMOVE REPETED STEPS FOR SAME CLIENT_ID

```python
# Sort the DataFrame by 'date_time' within each group
filtered_df_sorted = filtered_df.sort_values(by='date_time')

# Step 1: Group the DataFrame by 'visit_id'
grouped = filtered_df_sorted.groupby('visit_id')

# Step 2: Define a function to filter each group
def filter_group(group):
    # Keep only the earliest occurrence of each step
    filtered_group = group.drop_duplicates(subset='process_step')
    # Check if the group contains all four required steps
    if set(filtered_group['process_step']) == {'step_1', 'step_2', 'step_3', 'confirm'}:
        return filtered_group
    else:
        return pd.DataFrame()  # Return an empty DataFrame if the group does not meet the criteria

# Step 3: Apply the filtering function to each group and concatenate the results
filtered_clients = grouped.apply(filter_group).reset_index(drop=True)

# Display the resulting DataFrame
filtered_clients
```

# Completion Rate: proportion of users who reach the final â€˜confirmâ€™ step.

```python
df_control = df_web_all_2[df_web_all_2['Variation'] == 'Control']
df_test = df_web_all_2[df_web_all_2['Variation'] == 'Test']
```

```python
df_control['date_time'] = pd.to_datetime(df_control['date_time'])
# Step 2: Calculate time spent for each step
df_control['Time_spent'] = df_control.groupby('visit_id')['date_time'].diff()
```

```python
df_test['date_time'] = pd.to_datetime(df_test['date_time'])
# Step 2: Calculate time spent for each step
df_test['Time_spent'] = df_test.groupby('visit_id')['date_time'].diff()
```

```python
## Completion Rate:
pivot_df_control = df_control.pivot_table(index='process_step', aggfunc='size').reset_index(name='count')

start_count = pivot_df_control.loc[pivot_df_control['process_step'] == 'start', 'count'].iloc[0]

pivot_df_control['proportion_to_start'] = pivot_df_control['count'] / start_count

pivot_df_control
```

```python
## Completion Rate:
pivot_df_test = df_test.pivot_table(index='process_step', aggfunc='size').reset_index(name='count')

start_count = pivot_df_test.loc[pivot_df_test['process_step'] == 'start', 'count'].iloc[0]

pivot_df_test['proportion_to_start'] = pivot_df_test['count'] / start_count

pivot_df_test
```

```python
# Add 'table' column to identify data source
pivot_df_control['table'] = 'Control'
pivot_df_test['table'] = 'Test'

# Concatenate control and test tables
combined_table = pd.concat([pivot_df_control, pivot_df_test])

# Define the order of the process steps
process_step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Convert 'process_step' to categorical data with the specified order
combined_table['process_step'] = pd.Categorical(combined_table['process_step'], categories=process_step_order, ordered=True)

# Sort the DataFrame based on the categorical order
combined_table.sort_values(by='process_step', inplace=True)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))

# Plot control data
plt.plot(combined_table.loc[combined_table['table'] == 'Control', 'process_step'], 
         combined_table.loc[combined_table['table'] == 'Control', 'proportion_to_start'], 
         marker='o', linestyle='-', label='Control')

# Plot test data
plt.plot(combined_table.loc[combined_table['table'] == 'Test', 'process_step'], 
         combined_table.loc[combined_table['table'] == 'Test', 'proportion_to_start'], 
         marker='o', linestyle='-', label='Test')

plt.xlabel('Process Step')
plt.ylabel('Proportion to Start Count')
plt.title('Proportion of Each Step Compared to Start Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Error Rates

** Here, we don't use filtered_df, but instead df_well_all_2, since filtered_df has no errors

```python
# Assuming df_web_all2 is your DataFrame
# Step 1: Sort the DataFrame
df_web_all_2.sort_values(by=['client_id', 'visitor_id', 'date_time'], inplace=True)

# Step 2: Identify instances of going back to a previous step
df_web_all_2['previous_step'] = df_web_all_2.groupby(['client_id', 'visitor_id'])['process_step'].shift(1)

# Count instances where the current step is a step back
error_counts = df_web_all_2[df_web_all_2['process_step'] < df_web_all_2['previous_step']].groupby(['client_id', 'visitor_id']).size().reset_index(name='error_count')

# Step 3: Calculate error rates
total_visits = df_web_all_2.groupby(['client_id', 'visitor_id']).size().reset_index(name='total_visits')

# Merge error counts with total visits
error_rates_df = pd.merge(total_visits, error_counts, on=['client_id', 'visitor_id'], how='left').fillna(0)

# Calculate error rates
error_rates_df['error_rate'] = error_rates_df['error_count'] / error_rates_df['total_visits']

# Display the DataFrame with error rates
display(error_rates_df.head())
```

```python

error_rates_df = pd.merge(error_rates_df, df_final_clients[['client_id', 'Variation']], on='client_id', how='left')
```

```python
error_rates_df
```

```python
# Group data by 'Variation' and calculate the sum of 'error_count' for each group
grouped_data = error_rates_df.groupby('Variation')['error_count'].sum()

# Create a bar plot
plt.figure(figsize=(8, 6))
grouped_data.plot(kind='bar', color=['blue', 'orange'])
plt.title('Comparison of Errors in Test vs Control')
plt.xlabel('Variation')
plt.ylabel('Total Errors')
plt.xticks(rotation=0)
plt.show()
```
